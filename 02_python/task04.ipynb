## 문제

부정한 입력으로도 이상 종료 하지 않는 나눗셈 전용 계산 툴 작성

### 사양

- 두 개의 숫자 입력, 이를 나눔
- 숫자 대신 문자 입력한 경우: 오류: 숫자를 입력하십시오
- 나누는 수가 0: 오류: 0으로 나눌 수 없음
- 계산이 성공적 완료 -> 결과 표시

```python
try:
    target_num = float(input("1つ目の数を入力してください"))
    division_num = float(input("2つ目の数を入力してください"))

    result = target_num / division_num
    print(f"計算結果: {result}")
except ValueError:
    print("エラー: 数値を入力してください")
except ZeroDivisionError:
    print("0で割ることはできせん")
except Exception as e:
    print(f"予期しないエラーが発生しました: {e}")


```
