## 문제

10년 후의 금액을 복리로 예측하는 프로그램

### 사양

- 원금(초기 투자액) 입력
- 연도(백분율) 입력
- 10년간 금액 계산
- 복리: 이자를 차기 원금에 통합하는 방식
    - 예시 (원금 1000엔, 연 이율 5%)
        - 1년 후: 1000 * 1.05 = 1050.0
        - 2년 후: 1050.0 * 1.05 = 1102.5
        - 3년 후: 1102.5 * 1.05 = 1157.625...

```python
principal = int(input("元金を入力してください: "))
compound_interest = int(input("年利を入力してください: "))

compound_amount = principal

for i in range(10):
    compound_amount = compound_amount * (1 + (compound_interest / 100))

print(f"10年後の金額: {compound_amount:.2f}")
```
