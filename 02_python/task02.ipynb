# 문제

税別値段入力 →　税金込み値段＋配送料計算ツール
세금별도 가격 입력 -> 세금 포함 가격 + 배송비 계산 툴

## 사양

- 세금 별도 가격 입력
- 배송비 포함 가격 계산
    - 소비세 포함 가격 계산
        - 소비세 10%
    - 배송비 합한 가격 계산
        - 부가세 포함 2000이상: 무료
        - 부가세 포함 2000미만: 350엔
- 결과 표시

```python
price = int(input("値段を入力してください: "))
price_with_tax = int(price * 1.1)

def print_delivery_fee(delivery_fee):
    if delivery_fee == 0:
        print("送料は無料です。")
    else:
        print("送料は" + str(delivery_fee) + "円です。")

if price_with_tax >= 2000:
    delivery_fee = 0
else:
    delivery_fee = 350

print_delivery_fee(delivery_fee)

total_price = str( price_with_tax + delivery_fee )
print("送料込税込価格は" + total_price + "円です。")
```
