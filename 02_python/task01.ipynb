## 自己紹介プログラム作成

- 名前、年齢を保存する変数を用意し、値を割り当てる
- 上記の変数を使用し、自己紹介を出力する

### 問題の論理分解

名前、年齢を入力する
自己紹介文を用意する
名前、年齢、自己紹介文を合わせる
自己紹介内容を出力する

### コンピューターが理解できる形式で表現

名前: キム入力
年齢: 31入力
自己紹介文章: 初めまして、XXXと申します。年齢はXX歳です。来年はXX歳になります。よろしくお願いします。

### 実装


```python
name = "kim"
age = 25
introduction = f"初めまして、{name}と申します。年齢は{age}歳です。来年は{age+1}になります。よろしくお願いします。"

print(introduction)
print("初めまして、",name,"と申します。年齢は",age,"歳です。来年は",age+1,"になります。よろしくお願いします。")
print("初めまして、" + name + "と申します。年齢は" + str(age) + "歳です。来年は" + str(age+1) + "になります。よろしくお願いします。")
print("初めまして、{}と申します。年齢は{}歳です。来年は{}になります。よろしくお願いします。".format(name, age, age+1))
```
