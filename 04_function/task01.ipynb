```python
def validate_username(username):
    """
    ユーザー名のバリデーションを行う関数
    """
    if not (5 <= len(username) <= 15):
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    
    if " " in username:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    
    if not username.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    
    return True

input_name = input("ユーザー名を入力してください: ")

if validate_username(input_name):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
