```python
words = []

import string
alphabet_counts = {char: 0 for char in string.ascii_lowercase}

while True:
    user_input = input("英単語を入力してください：").strip()
    
    if user_input == "":
        break
    
    words.append(user_input)

    for char in user_input.lower():
        if char in alphabet_counts:
            alphabet_counts[char] += 1

words.sort()

print(f"入力した英単語：{words}")

for char, count in alphabet_counts.items():
    if count > 0:
        print(f"{char}が{count}個ありました")
```
